```python
import numpy as np
import matplotlib.pyplot as plt
NUM_SAMPLES = 100
NOISE = 0

def gen_spiral(delta_t, label):
    data_points = []
    label_points = []
    n = int(NUM_SAMPLES/2)
    for i in range(1, n+1):
      r = i / n * 5
      t = 1.75 * i / n * 2 * np.pi + delta_t
      x = r * np.sin(t) + np.random.uniform(-1, 1) * NOISE
      y = r * np.cos(t) + np.random.uniform(-1, 1) * NOISE
      data_points.append([x, y])
      label_points.append(label)
    return np.array(data_points), np.array(label_points)
```

```python
X1, Y1 = gen_spiral(0, 1)
X2, Y2 = gen_spiral(np.pi, -1)
X = np.concatenate((X1, X2), axis=0)
Y = np.concatenate((Y1, Y2), axis=0)
```

```python
R = np.hypot(X[:,0], X[:,1])
THETA = np.arctan2(X[:,0], X[:,1])
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1], c=Y, cmap='bwr')
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.savefig("spiral.pdf")
plt.show()
```

```python
plt.scatter(R, THETA, c=Y, cmap='bwr')
plt.xlabel(r"$r = \sqrt{x_1^2 + x_2^2}$")
plt.ylabel(r"$\theta = \arctan2(x_2, x_1)$")
plt.savefig("spiral_rtheta.pdf")
plt.show()
```

```python
plt.scatter(R, np.mod(THETA - 2 * R, 2 * np.pi), c=Y, cmap='bwr')
plt.xlabel(r"$r = \sqrt{x_1^2 + x_2^2}$")
plt.ylabel(r"$(\theta - 2r)~\mathrm{mod}~2\pi$")
plt.savefig("spiral_rtheta_mod.pdf")
plt.show()
```
